```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


# Make numpy printouts easier to read.
np.set_printoptions(precision=3, suppress=True)
```

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing

print(tf.__version__)
```

## Dataset Importing

```python
url = 'covid_data.xlsx'
column_names = ['Country (label)', 'Date (label)', 'new_cases', 'total_cases', 'new_deaths', 'total_deaths', 'new_tests', 'total_tests', 'stringency_index', 'population', 'population_density', 'median_age', 'aged_65_older', 'gdp_per_capita', 'cardiovasc_death_rate', 'diabetes_prevalence', 'human_development_index', 'cluster']

raw_dataset = pd.read_excel(url, sheet_name='covid_data', index=False)
```

```python
dataset = raw_dataset.copy()
CLUSTER_NUMBER = dataset["cluster"].max() + 1
print(CLUSTER_NUMBER)
dataset.tail()
```

```python
dataset.isna().sum()
```

```python
dataset = dataset.dropna()
dataset.isna().sum()
```

```python
dataset.pop('total_deaths')
dataset.tail()
```

```python
split_by_cluster = [dataset.where(dataset['cluster'] == i) for i in range(0,CLUSTER_NUMBER)]
split_by_cluster = [split_by_cluster[i].dropna() for i in range(0,CLUSTER_NUMBER)]

for i in range(0, CLUSTER_NUMBER):
    split_by_cluster[i] = split_by_cluster[i].sort_values(by=['Country (label)'])
    
for i in range(0, CLUSTER_NUMBER):
    split_by_cluster[i] = split_by_cluster[i].drop(columns=['Unnamed: 0'])
    
split_by_cluster[0]
```

```python
split_by_country = []
for j in range(0, CLUSTER_NUMBER):
    split_by_country.append([])
for j in range(0, CLUSTER_NUMBER):
    for i in range(1, 210):
        df = split_by_cluster[j].where(dataset['Country (label)'] == i) 
        if not df.empty:
            split_by_country[j].append(df.dropna())
```

```python
for i in range(0, CLUSTER_NUMBER):
    sub_list = split_by_country[i]
    indices = []
    for j in range(len(sub_list)):
        if not sub_list[j].empty:
            indices.append(j)
    split_by_country[i] = [sub_list[index] for index in indices]
```

```python
TRAIN_PERCENT = .7
VALIDATION_PERCENT = .15
country_train_dataset = []
country_validation_dataset = []
country_test_dataset = []

for i in range(0, CLUSTER_NUMBER):
    train_split_index = round(TRAIN_PERCENT * len(split_by_country[i]))
    valid_split_index = round((TRAIN_PERCENT + VALIDATION_PERCENT) * len(split_by_country[i]))
    np.random.seed(5)    
    np.random.shuffle(split_by_country[i])
    country_train_dataset.append(split_by_country[i][0:train_split_index])
    country_validation_dataset.append(split_by_country[i][train_split_index:valid_split_index])
    country_test_dataset.append(split_by_country[i][valid_split_index:len(split_by_country[i])])
```

```python
train_dataset = []
validation_dataset = []
test_dataset = []

for i in range(0, CLUSTER_NUMBER):
    train_dataset.append(pd.concat(country_train_dataset[i]))
    validation_dataset.append(pd.concat(country_validation_dataset[i]))
    test_dataset.append(pd.concat(country_test_dataset[i]))
```

```python
for i in range(0, CLUSTER_NUMBER):
    print(pd.unique(sorted(train_dataset[i]['Country (label)'])))
```

```python
for i in range(0, CLUSTER_NUMBER):
    print(pd.unique(sorted(validation_dataset[i]['Country (label)'])))
```

```python
for i in range(0, CLUSTER_NUMBER):
    print(pd.unique(sorted(test_dataset[i]['Country (label)'])))
```

# Visual Analysis 

```python
def plot_covid_deaths(data, country_label, index):
    country_data = data.where(data['Country (label)'] == country_label)
    country_data = country_data.dropna()
    plt.figure(index+1)
    plt.plot(country_data['Date (label)'], country_data['new_deaths'])
    plt.xlabel('Days since 4/1/2020')
    plt.ylabel('new_deaths')
    plt.title('Country {}'.format(country_label))
    plt.grid(True)
```

```python
for i in range(0, CLUSTER_NUMBER):
    plot_covid_deaths(train_dataset[i], 28, i)
```

```python
for i in range(0, CLUSTER_NUMBER):
    sns.pairplot(train_dataset[i][['Date (label)', 'new_cases', 'new_deaths', 'new_tests']], diag_kind='kde')
```

```python
for i in range(0, CLUSTER_NUMBER):
    sns.pairplot(train_dataset[i][['new_deaths', 'stringency_index', 'population_density', 'gdp_per_capita']], diag_kind='kde')
```

```python
train_dataset[0].describe().transpose()
```

```python
train_features = []
validation_features = []
test_features = []

train_labels = []
validation_labels = []
test_labels = []

for i in range(0, CLUSTER_NUMBER):
    train_features.append([])
    validation_features.append([])
    test_features.append([])
    train_labels.append([])
    validation_labels.append([])
    test_labels.append([])
    
for j in range(0, CLUSTER_NUMBER):
    train_features[j] = train_dataset[j].copy()
    validation_features[j] = validation_dataset[j].copy()
    test_features[j] = test_dataset[j].copy()

for k in range(0, CLUSTER_NUMBER):
    train_labels[k] = train_features[k].pop('new_deaths')
    validation_labels[k] = validation_features[k].pop('new_deaths')
    test_labels[k] = test_features[k].pop('new_deaths')
```

```python
train_dataset[0].describe().transpose()[['mean', 'std']]
```

## Normalization

```python
normalizer = []

for i in range(0, CLUSTER_NUMBER):
    normalizer.append([])
    normalizer[i] = preprocessing.Normalization()
```

```python
for i in range(0, CLUSTER_NUMBER):
    temp = np.array(train_features[i])[:, 0:-1]
    normalizer[i].adapt(temp)
```

```python
for i in range(0, CLUSTER_NUMBER):
    print(normalizer[i].mean.numpy())
```

```python
first = np.array(train_features[0])[:1, 0:-1]

with np.printoptions(precision=2, suppress=True):
    print('First example:', first)
    print()
    print('Normalized:', normalizer[2](first).numpy())
```

## Linear Regression

```python
linear_model = []

for i in range(0, CLUSTER_NUMBER):
    linear_model.append([])
    linear_model[i] = tf.keras.Sequential([normalizer[i], layers.Dense(units=1)])
```

```python
for i in range(0, CLUSTER_NUMBER):
    print(linear_model[i].predict(np.array(train_features[i])[:10, 0:-1]))
```

```python
linear_model[0].layers[1].kernel
```

```python
for i in range(0, CLUSTER_NUMBER):
    linear_model[i].compile(
        optimizer=tf.optimizers.Adam(learning_rate=0.1),
        loss='mean_absolute_error')
```

```python
%%time
history = []
for i in range(0, CLUSTER_NUMBER):
    history.append([])
    
for i in range(0, CLUSTER_NUMBER):
    train_features_temp = np.array(train_features[i])[:, 0:-1]
    train_labels_temp = np.array(train_labels[i])
    validation_features_temp = np.array(validation_features[i])[:, 0:-1]
    validation_labels_temp = np.array(validation_labels[i])
    history[i] = linear_model[i].fit(train_features_temp, train_labels_temp, epochs = 100, verbose = 0,
                                    validation_data = (validation_features_temp,validation_labels_temp))
```

```python
def plot_loss(history, model_str, index):
    plt.figure(index+1)
    plt.plot(history.history['loss'], label='loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.ylim([0, 30])
    plt.xlabel('Epoch')
    plt.ylabel('Error [new_deaths]')
    plt.title('{} Mean Absolute Error'.format(model_str))
    plt.legend()
    plt.grid(True)
```

```python
for i in range(0, CLUSTER_NUMBER):
    plot_loss(history[i], 'Linear Regression', i)
```

```python
test_results = []

for i in range(0, CLUSTER_NUMBER):
    test_results.append({})
    test_results[i]['linear_model'] = linear_model[i].evaluate(np.array(test_features[i])[:, 0:-1], np.array(test_labels[i]), verbose=0)
```

## DNN Model

```python
def build_and_compile_model(norm):
    model = keras.Sequential([
        norm,
        layers.Dense(64, activation='relu'),
        layers.Dense(64, activation='relu'),
        layers.Dense(1)
    ])

    model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(0.001))
    return model
```

```python
dnn_model = []

for i in range(0, CLUSTER_NUMBER):
    dnn_model.append([])
    dnn_model[i] = build_and_compile_model(normalizer[i])

dnn_model[0].summary()
```

```python
%%time
history = []

for i in range(0, CLUSTER_NUMBER):
    history.append([])
    
for i in range(0, CLUSTER_NUMBER):
    train_features_temp = np.array(train_features[i])[:, 0:-1]
    train_labels_temp = np.array(train_labels[i])
    validation_features_temp = np.array(validation_features[i])[:, 0:-1]
    validation_labels_temp = np.array(validation_labels[i])
    history[i] = dnn_model[i].fit(train_features_temp, train_labels_temp, epochs = 100, verbose = 0,
                                    validation_data = (validation_features_temp,validation_labels_temp))
```

```python
for i in range(0, CLUSTER_NUMBER):
    plot_loss(history[i], 'DNN', i)
```

```python
for i in range(0, CLUSTER_NUMBER):
    test_results[i]['dnn_model'] = dnn_model[i].evaluate(np.array(test_features[i])[:, 0:-1], np.array(test_labels[i]), verbose=0)
```

## Performance Analysis

```python
linear_test_predictions = []

for i in range(0, CLUSTER_NUMBER):
    linear_test_predictions.append([])
    linear_test_predictions[i] = linear_model[i].predict(np.array(test_features[i])[:, 0:-1]).flatten()

for i in range(0, CLUSTER_NUMBER):
    plt.figure(i+1)
    plt.scatter(test_labels[i], linear_test_predictions[i])
    plt.xlabel('True Values [new_deaths]')
    plt.ylabel('Predictions [new_deaths]')
    plt.title('Linear Regression True VS Predicted')
    lims = [0, 500]
    plt.xlim(lims)
    plt.ylim(lims)
    _ = plt.plot(lims, lims)
```

```python
dnn_test_predictions = []

for i in range(0, CLUSTER_NUMBER):
    dnn_test_predictions.append([])
    dnn_test_predictions[i] = dnn_model[i].predict(np.array(test_features[i])[:, 0:-1]).flatten()

for i in range(0, CLUSTER_NUMBER):
    plt.figure(i)
    plt.scatter(test_labels[i], dnn_test_predictions[i])
    plt.xlabel('True Values [new_deaths]')
    plt.ylabel('Predictions [new_deaths]')
    plt.title('DNN True VS Predicted')
    lims = [0, 500]
    plt.xlim(lims)
    plt.ylim(lims)
    _ = plt.plot(lims, lims)
```

```python
error = []

for i in range(0, CLUSTER_NUMBER):
    error.append([])
    error[i] = linear_test_predictions[i] - test_labels[i]

for i in range(0, CLUSTER_NUMBER):
    plt.figure(i)
    plt.hist(error[i], bins=25)
    plt.xlabel('Prediction Error [new_deaths]')
    _ = plt.ylabel('Count')
    plt.title('Linear Regression Prediction Error')
```

```python
error = []

for i in range(0, CLUSTER_NUMBER):
    error.append([])
    error[i] = dnn_test_predictions[i] - test_labels[i]

for i in range(0, CLUSTER_NUMBER):
    plt.figure(i)
    plt.hist(error[i], bins=25)
    plt.xlabel('Prediction Error [new_deaths]')
    _ = plt.ylabel('Count')
    plt.title('DNN Prediction Error')
```

```python
def predicted_covid_deaths_then_plot(model, test_dataset, country_id, count):
    test_dataset = test_dataset.where(test_dataset['Country (label)'] == country_id)
    test_dataset = test_dataset.dropna()

    test_features = test_dataset.copy()
    test_labels = test_features.pop('new_deaths')
    
    test_predictions = model.predict(test_features).flatten()
    test_predictions = [round(test_predictions[i]) for i in range(len(test_predictions))]
    
    plt.figure(count)
    plt.plot(test_dataset['Date (label)'], test_dataset['new_deaths'], label='True Values')
    plt.plot(test_dataset['Date (label)'], test_predictions, label='Predicted Values')

    plt.xlabel('Days since 4/1/2020')
    plt.ylabel('new_deaths')
    plt.title('Country {}'.format(country_id))
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
count = 1
for j in range(0, CLUSTER_NUMBER):
    for i in pd.unique(sorted(test_dataset[j]['Country (label)'])):
        predicted_covid_deaths_then_plot(linear_model[j], test_dataset[j].iloc[:, 0:-1], i, count)
        count = count+1
```

```python
count = 1
for j in range(0, CLUSTER_NUMBER):
    for i in pd.unique(sorted(test_dataset[j]['Country (label)'])):
        predicted_covid_deaths_then_plot(dnn_model[j], test_dataset[j].iloc[:, 0:-1], i, count)
        count = count+1
```
