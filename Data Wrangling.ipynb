```python
import pandas as pd
import pycountry

locations_mapping_file = 'IHME_GBD/IHME_GBD_2019_ALL_LOCATIONS_HIERARCHIES_Y2020M10D15.XLSX'
locations_mapping = pd.read_excel(locations_mapping_file)[['Location ID', 'Location Name']]

locations_mapping
```

```python
cause_mapping_file = 'IHME_GBD/IHME_GBD_2019_CAUSE_HIERARCHY_Y2020M10D15.XLSX'
cause_mapping = pd.read_excel(cause_mapping_file)[['Cause ID', 'Cause Name']]

cause_mapping
```

```python
diseases_prevelance_file = 'IHME_GBD/prevelance_diseases.csv'
diseases_data = pd.read_csv(diseases_prevelance_file, sep=',')

diseases_data
```

```python
# Join on location and cause

disease_country = diseases_data.join(locations_mapping.set_index('Location ID'), on='location').drop_duplicates()
disease_country

disease_country = disease_country.join(cause_mapping.set_index('Cause ID'), on='cause')[['Location Name', 'Cause Name', 'val']].drop_duplicates()
disease_country = disease_country.fillna('Total Cancers')
```

```python
disease_country
```

```python
df = disease_country.pivot(index='Location Name', columns='Cause Name')
df.columns = ['_'.join((col[1], col[0])) for col in df.columns]
```

```python
print(df.columns)
df
```

```python
def country_lookup(x):
    try:
        return pycountry.countries.lookup(x).alpha_3
    except:
        return 'Unknown'
    
df['iso'] = df.index
df['iso'] = df['iso'].apply(country_lookup)
df
```

```python
df.to_excel('disease_prevalence_per_country.xlsx')
```
